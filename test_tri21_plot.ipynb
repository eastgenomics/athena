```python
import pandas as pd

import pandas as pd
import matplotlib as plt

import plotly as plt

pd.options.plotting.backend = "plotly"
import plotly.io as pio
pio.renderers.default = "browser"
import plotly.graph_objects as go
```

```python
df = pd.read_csv('1295_0262.csv')
```

```python
set(df[df['gene'] == 'U2AF1']['exon'].tolist())
```

```python

import re
import sys
fig = go.Figure()

# df = df[df['chrom'] == '21' ]

df2 = df[df['gene'] == 'MSH2']

# df2 = df2[df2['exon'] == 1]

# df2 = df2[[x for x in df2.columns if x != 'S26']]


cols = df2.columns.tolist()

cols = [x for x in cols if re.match('S[\d]+_[norm|raw]', x)]
# cols.append('S0_raw')


mean_cols  = [x for x in cols if re.match('S[\d]+_norm', x)]

print(cols)
# cols = [x for x in cols if x.startswith('S0')]

# sys.exit(0)

df2['mean'] = df2[mean_cols].mean(axis=1).round()
df2['std'] = df2[mean_cols].std(axis=1)
df2['std_+3'] = df2['mean'] + df2['std'] * 2
df2['std_-3'] = df2['mean'] - df2['std'] * 2


fig.add_trace(go.Scatter(x=df2['position'], y=df2['std_-3'], hoverinfo='skip',
    mode='lines', name='-3 std dev',line=dict(color='rgba(197, 243, 243, 0.85)')))
fig.add_trace(go.Scatter(x=df2['position'], y=df2['std_+3'], fillcolor='rgba(197, 243, 243, 0.85)', hoverinfo='skip',
    mode='lines', name='+3 std dev', fill='tonexty', line=dict(color='rgba(197, 243, 243, 1)')))
fig.add_trace(go.Scatter(x=df2['position'], y=df2['mean'], mode='lines', name='mean', line=dict(color='skyblue')))

for col in cols:
    fig.add_trace(go.Scatter(x=df2['position'], y=df2[col],
        mode='lines', name=col))

df2.columns


fig.update_layout(hovermode="closest")
fig.show()
```
