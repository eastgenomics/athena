```python
hsmetrics_paths = glob.glob('/home/jethro/Projects/athena/test/tri21/hs/*')
hsmetrics = {}

for file in hsmetrics_paths:
    name = Path(file).stem.replace('_markdup.hsmetrics', '')

    df = load.LoadData().read_hsmetrics(file)
    df.columns = df.iloc[0].tolist()
    df.reset_index(inplace=True, drop=True)
    df.drop(index=df.index[0], axis=0, inplace=True)
    hsmetrics[name] = df
```

```python
all_hsmetrics = pd.concat(hsmetrics.values())
all_hsmetrics
all_hsmetrics = all_hsmetrics.astype({'ON_TARGET_BASES': int, 'PCT_USABLE_BASES_ON_TARGET': float, 'TOTAL_READS': int})
# normalisation_value = (all_hsmetrics['ON_TARGET_BASES'] *\
#                  all_hsmetrics['PCT_USABLE_BASES_ON_TARGET']).sum()
normalisation_value = all_hsmetrics['TOTAL_READS'].sum()
```

```python
count = 1

all_dfs = None
all_dfs = list(dfs.values())[0].drop(['S1'], axis=1)


for name, bed in dfs.items():
    # print(df)
    # bases = hsmetrics[name]['ON_TARGET_BASES']
    # pct = hsmetrics[name]['PCT_USABLE_BASES_ON_TARGET']
    
    df = bed.copy()
    hsmetrics[name] = hsmetrics[name].astype({
            'ON_TARGET_BASES': int,
            'PCT_USABLE_BASES_ON_TARGET': float,
            'TOTAL_READS': int
        })

    # sample_bases = (
    #         hsmetrics[name]['ON_TARGET_BASES'] * hsmetrics[name]['PCT_USABLE_BASES_ON_TARGET']
    #     ).iloc[0] / normalisation_value

    sample_bases = hsmetrics[name]['TOTAL_READS'].iloc[0] / normalisation_value

    df[f'S{count}_norm'] = df[f"S{count}"] * sample_bases

    all_dfs[f"S{count}"] = df[f"S{count}"]
    all_dfs[f'S{count}_norm'] = df[f'S{count}_norm']

    # print(name)
    # print(df[f'S{count}'].mean())
    # print(df[f'S{count}_norm'].mean())
    # print(' ')
    count += 1

print(all_dfs)



```

```python
def is_numeric(value:str) -> bool:
    """
    """
    if not value:
        return value
    try:
        if str(value)[0] == '0' and len(value) > 1 and not str(value)[1] == '.':
            # catch things such as 01 which aren't real numbers
            return value
    except Exception:
        # could raise lots of errors (i.e IndexErrors from slicing)
        # lazily catch all and continue
        return value

    try:
        float(value)
        return True
    except ValueError:
        return value
    
    # it can float, check if it's an int or float
    if '.' in str(value):
        return float(value)
    else:
        return int(value)
```
